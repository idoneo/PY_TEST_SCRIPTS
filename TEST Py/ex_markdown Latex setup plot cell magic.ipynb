##Simple Plotting Example

This is a plot of the $sinc$ function

$$
f(x) = \frac{\sin(x)}{x}
$$

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
pi= np.pi
x = np.linspace(-4*pi,4*pi,1000)

plt.plot(x,np.sin(x)/x)
plt.show()
```

```python
#switch to vector graphic magic
#add xtick to axis
#make plot size larger
#increase font
#add label and legend smll r = rich text $$ = Latex
% matplotlib inline
%config InlineBackend.figure_format = 'svg'
pi= np.pi
x = np.linspace(-4*pi,4*pi,1000)
plt.rcParams ['figure.figsize'] = (10,4)
plt.rcParams.update({'font.size':17})
plt.yticks([0,1],['$0$','$1$'])
plt.xticks([-4*pi,-2*pi,0, 2*pi, 4*pi],
          ['$-4\pi$','$-2\pi$', '$0$' ,'$2\pi$','$4\pi$'])
plt.axhline(0,color='blue', lw =1)
plt.plot(x,np.sin(x)/x, label = r'$f(x) =\frac{\sin(x)}{x}$')
plt.ylim(-0.5,1.5)
plt.title ('The $sinc$ Function')
plt.legend(loc='best', fontsize = 20)
plt.show()
```
