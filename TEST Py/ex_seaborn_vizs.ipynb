```python
iris = sns.load_dataset('iris')
```

```python
iris.head(2)
```

```python
# scatterplot with one categorical variable
viz1 = sns.stripplot(x='species',y='petal_length',data =iris)
```

```python
#categorical scatterplot with non-overlapping ponts
viz2 = sns.swarmplot(x='species',y='petal_length',data =iris)
```

```python
#boxplot with wideform data
viz3 = sns.boxplot(data= iris,orient='h')
```

```python
#plot data and regression model fits across a facet grid
viz4 = sns.lmplot(x='sepal_width',y ='sepal_length',hue= 'species',data= iris)
```

```python
#plot a bivariate distribution

sns.jointplot('sepal_length','sepal_width',data=iris,kind='kde')
```

```python
sns.regplot('sepal_length','sepal_width',data=iris)
```
